## Part 2 - 가설 검정

### Hypothesis Test    
   - 수치형 (numerical/continuous) 칼럼이 어떤 차이를 보이는지 t-test를 통해 가설을 세우고 검증하세요. (최소 2개)
   - 범주형 (categorical) 칼럼이 어떤 차이를 보이는지 $\chi^2-test$를 통해 가설을 세우고 검증하세요. (최소 2개)  
  **Hint:** categorical feature 간의 "cross tab"을 생성하여 탐색해 보세요 (aka [contingency tables](https://en.wikipedia.org/wiki/Contingency_table)). 그리고 $\chi^2 test$를 적용해 보세요. [pandas.crosstab](http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.crosstab.html)을 사용하면 contingency table 을 생성할 수 있습니다. 그리고 [scipy.stats.chi2_contingency](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html)를 사용하면 $\chi^2$ statistics 를 계산할 수 있습니다.    

```python
# 과제 2 - 1 - 1: 수치형 칼럼 t-test 시행 (정육 - 어류)

import scipy.stats
import numpy as np
m = []

for i in range(10):
  val1 = df.MntMeatProducts.sample(n=30,replace=True)
  val2 = df.MntFishProducts.sample(n=30,replace=True)
  value = scipy.stats.ttest_rel(val1, val2)
  m.append(value[1])
  #print(value[1])

print('P-value :',np.mean(m))
```

```python
# 과제 2 - 1 - 2: 수치형 칼럼 t-test 시행 (온라인 스토어 구매 - 오프라인 스토어 구매)
m = []

for i in range(10):
  val1 = df.NumWebPurchases.sample(n=30,replace=True)
  val2 = df.NumStorePurchases.sample(n=30,replace=True)
  value = scipy.stats.ttest_rel(val1, val2)
  m.append(value[1])
  #print(value[1])

print('P-value :',np.mean(m))
```

```python
# 과제 2 - 2 - 1 : 범주형 칼럼  카이스퀘어 검정 시행 (교육수준 - 결혼상태)
import pandas as pd
from scipy.stats import chi2_contingency

df1 = pd.crosstab(df.Education,df.Marital_Status)
chiresult = chi2_contingency(df1, correction=False)
print('Chi square: {}'.format(chiresult[0]))
print('P-value: {}'.format(chiresult[1]))
print('Degrees of freedom: {}'.format(chiresult[2]))
df1
```

```python
# 과제 2 - 2 - 2 : 범주형 칼럼  카이스퀘어 검정 시행 (국가 - 교육수준)

df2 = pd.crosstab(df.Country,df.Education)
chiresult = chi2_contingency(df2, correction=False)
print('Chi square: {}'.format(chiresult[0]))
print('P-value: {}'.format(chiresult[1]))
print('Degrees of freedom: {}'.format(chiresult[2]))
df2
```

1.   t-test 해석
    1. 정육 - 어류 분석
      - H0 = 정육 과 어류 간 평균 차이는 없을 것이다.
      - H1 = 정육 과 어류 간 평균 차이가 있을 것이다.
      - P-value < 0.05 이므로 H0 기각 =  정육 판매량에 따른 어류 판매량의 차이는 있다.

    2. 온라인 스토어 판매 - 오프라인 스토어 판매 분석
      - H0 = 온라인 과 오프라인 간 평균 차이는 없을 것이다.
      - H1 = 온라인 과 오프라인 간 평균 차이가 있을 것이다.
      - P-value > 0.05 이므로 H0를 기각할 수 없다. =  온라인과 판매량에 따른 오프라인 판매량의 차이는 없다.


2.   χ2  test 해석

    1. 교육수준 - 결혼상태 분석
      - H1 = 교육수준과 결혼상태는 연관관계가 있을 것이다.
      - h0 = 교육수준과 결혼상태는 연관관계가 없다.
      - P-value > 0.05 이므로 H0 기각 실패 =  교육수준과 결혼상태 사이의 관계는 없다고 결론
      
    2. 국가 - 교육수준 분석
      - H1 = 국가와 교육수준은 연관관계가 있을 것이다.
      - H0 = 국가와 교육수준은 연관관계가 없다.
      - P-value > 0.05 이므로 H0 기각 실패 = 국가와 교육수준은 독립적이라고 결론
